```python
# Import all necessary libraries
import pandas as pd                 # For data manipulation
import numpy as np                  # For numerical computations
import seaborn as sns               # For advanced visualizations
import matplotlib.pyplot as plt     # For basic plotting
import warnings

# Settings to improve readability
warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)

# Load the dataset
file_path="./mental_health_workplace_survey.csv"
data=pd.read_csv(file_path)  

# Preview first few rows
data.head()
```

```python
# Shape of the dataset
print("Total Rows:", data.shape[0])
print("Total Columns:", data.shape[1])

# Check data types and non-null counts
data.info()

# Check summary statistics for numeric columns
data.describe()
```

```python
# Check total missing values per column
data.isnull().sum().sort_values(ascending=False)
```

```python
# Handle numeric columns (int/float)
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    median_val = data[col].median()
    data[col].fillna(median_val, inplace=True)  # Replaces missing values with median
    print(f"Filled missing values in {col} with median: {median_val}")

# Handle categorical columns (object/string)
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    mode_val = data[col].mode()[0]
    data[col].fillna(mode_val, inplace=True)    # Replaces missing values with most frequent value
    print(f"Filled missing values in {col} with mode: {mode_val}")
```

```python
# Plot boxplots for key numeric features
features = ['WorkHoursPerWeek', 'SleepHours', 'StressLevel']
for col in features:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(data=data, x=col)
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
def remove_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    original_size = data.shape[0]
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    removed = original_size - data.shape[0]
    print(f"{removed} outliers removed from '{col}'")
    return data

# Remove outliers from important features
for col in ['WorkHoursPerWeek', 'SleepHours', 'StressLevel']:
    data = remove_outliers_iqr(data, col)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="EmployeeID", y="BurnoutLevel", hue="BurnoutLevel",data=data)
plt.title("Burnout Level")
plt.xlabel("Employee ID")
plt.ylabel("Burnout Level")
plt.show()
```

```python
# Scatterplot of Stress vs Sleep colored by Burnout
sns.scatterplot(x='SleepHours', y='StressLevel', hue='BurnoutLevel', data=data, palette='coolwarm')
plt.title("Stress vs Sleep (Colored by Burnout)")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.legend(title="Burnout")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=data, x="HasMentalHealthSupport", y="BurnoutLevel", palette="Set2")
plt.title("Burnout Level by Access to Mental Health Support")
plt.xlabel("Has Mental Health Support?")
plt.ylabel("Burnout Level")
plt.tight_layout()
plt.show()
```
